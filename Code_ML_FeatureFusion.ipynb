```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
demoji.download_codes()
def emo(text):
  temp=emoji.demojize(str(text),delimiters=(" "," "))
  temp=temp.replace("_","  ")
  return temp
string.punctuation
punct=!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~“, ' ’’“
def clean_text(text):
    text = re.sub('@[a-zA-Z0-9]*','',text)
    text=contractions.fix(text,slang=True)
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    text="".join([word.lower() for word in text if word not in punct])
    text = "".join([word for word in text if not word.isdigit()])
    #text = " ".join(word for word in text.split() if word not in stopwords)
    tokens = word_tokenize(text)
    text = " ".join(lemmatizer.lemmatize(word.lower()) for word in tokens)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vec = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), max_df=1.0, min_df=1, max_features=5000)
count_train = Tfidf_vec.fit(df_train['clean_text'])
train_features1 = Tfidf_vec.transform(df_train['clean_text'])
val_features1 = Tfidf_vec.transform(df_val['clean_text'])
test_features1 = Tfidf_vec.transform(df_test['clean_text'])
```

```python
from sklearn.feature_extraction.text import  CountVectorizer, TfidfTransformer,TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(analyzer='char',ngram_range=(1,3))
tfidf_vectorizer.fit_transform(df_train['translit_text'])
train_features2=  tfidf_vectorizer.transform(df_train['translit_text'])
val_features2=  tfidf_vectorizer.transform(df_val['translit_text'])
test_features2 =  tfidf_vectorizer.transform(df_test['translit_text'])
```
